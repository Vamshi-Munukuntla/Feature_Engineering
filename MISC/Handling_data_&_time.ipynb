# Working with Dates

```python
df_date.date = pd.to_datetime(df_date.date)
```

```python
df_date.info()
```

#### 1. Extract year

```python
df_date['year'] = df_date.date.dt.year
```

```python
df_date.head()
```

```python
df_date.year.unique()
```

#### 2. Extract Month

```python
df_date['month'] = df_date.date.dt.month
df_date.head()
```

```python
df_date['month_name'] = df_date.date.dt.month_name()
df_date.head()
```

#### Extract Days

```python
df_date['days'] = df_date.date.dt.day
df_date.head()
```

```python
# Day of week
df_date['day_of_week'] = df_date.date.dt.day_of_week
df_date.head()
```

```python
df_date.day_of_week.unique()
```

```python
df_date['Week'] = df_date.date.dt.week
df_date.head()
```

```python
df_date['day_name'] = df_date.date.dt.day_name()
df_date.head()
```

```python
# is weekend?

df_date['date_is_weekend'] = np.where(df_date.day_name.isin(['Sunday','Saturday']),1,0)
df_date.head()
```

```python
df_date.date_is_weekend.value_counts()
```

## Extract Quarter

```python
df_date['Quarter'] = df_date.date.dt.quarter
df_date.head()
```

## Extract Semester

```python
df_date['semester'] = np.where(df_date.Quarter.isin([1,2]),1,2)
df_date.head()
```

## Extract Time elapsed between dates

```python
import datetime
```

```python
today = datetime.datetime.today()
today
```

```python
today - df_date.date
```

```python
(today - df_date.date).dt.days
```

```python
# Months Passed

np.round((today - df_date.date)/np.timedelta64(1,'M'),0)
```
