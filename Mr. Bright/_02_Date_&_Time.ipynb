### Extract Weekday from DateTime

Extracting Weekday is very helpful in feature engineering. 

For instance, the target variable can be dependent on the day of the week, like sales of a product are generally higher on a weekend.

The **weekday()** method returns an integer value for the day of the week.

Monday : 0 & Sunday : 6



We can also use the **isoweekday()** to return value between 1 and 7, just like in real world.


```python
date = datetime.datetime.now()
date
```

```python
date.weekday() # Wednesday : 2
```

```python
date.isoweekday() # Wednesday : 3
```

```python
date=datetime.datetime.now()
# week starts from 0
print('Today is Wednesday and it is representedby weekday() as:',
      date.weekday()) # output 2 for Wednesday

# # week starts with 1
print('Today is Wednesday and it is represented by isoweekday() as:',
      date.isoweekday()) # output 3 in ISO format for Wednesday
```

### DateTime Formatting with Strptime and Strftime

The way date & time is represented may be different in different places.

It's more common to use `mm/dd/yyyy` in the **US**,
whereas `dd/mm/yyyy` is more common in the **UK**.

Python has **strftime()** & **strptime()** methods to handle this.

### Python **strftime()** : Converts datetime object to string
The method creates a formatted string from a given date, datetime or time object.

```python
import datetime

# current date and time
now = datetime.datetime.now()
print('date and time:', now)

t = now.strftime('%H:%M:%S')
print('time :', t)

s1 = now.strftime('%m/%d/%y, %H:%M:%S') # US FORMAT
# mm/dd/YY H:M:S format
print('USA',s1)

s2 = now.strftime('%d/%m/%y, %H:%M:%S') # US FORMAT
# mm/dd/YY H:M:S format
print('UK ',s2)
```

Here, %Y, %m, %d, %H etc. are format codes. The strftime() method takes one or more format codes and returns a formatted string based on it.

In the above program, t, s1 and s2 are strings.

1. **%Y** - year [0001,..., 2018, 2019,..., 9999]
2. **%m** - month [01, 02, ..., 11, 12]
3. **%d** - day [01, 02, ..., 30, 31]
4. **%H** - hour [00, 01, ..., 22, 23
4. **%M** - minute [00, 01, ..., 58, 59]
6. **%S** - second [00, 01, ..., 58, 59]

### Extract the Year, Month, Day, and Time using **strftime()**

```python
now = datetime.datetime.now()
print(now)
```

```python
year = now.strftime("%Y")
print('YEAR :',year)

month = now.strftime("%m")
print('MONTH :',month)

day = now.strftime('%d')
print('DAY :',day)

time = now.strftime('%H:%M:%S')
print('TIME :',time)

date_time = now.strftime("%m/%d/%Y, %H:%M:%S")
print("DATE & TIME:",date_time)
```

Here, **year**, **day**, **time** and **date_time** are strings, whereas **now** is a datetime object.

### Creating a string from timestamp

```python
import datetime
```

```python
timestamp = 8897596545
date_time = datetime.datetime.fromtimestamp(timestamp)

print("Date time object :",date_time)

d = date_time.strftime("%d/%m/%Y, %H:%M:%S")
print("Output 2:", d)

d = date_time.strftime('%d %b %Y')
print("Output 3:", d)

d = date_time.strftime('%d %B %Y')
print("Output 4:", d)

d = date_time.strftime('%I %p')
print("Output 5:", d)
```

### Python  **strftime()** : converts string to datetime

The strptime() method creates a datetime object from a given string (representing date and time).

```python
import datetime

date_string = '21 May, 1995'
print('date_string :',date_string)

date_object = datetime.datetime.strptime(date_string, '%d %B, %Y')
print(date_object)
```

By the way, %d, %B and %Y format codes are used for day, month(full name) and year respectively.

%d : day

%B : month (full name)

### Get year, month, and day from a DataFrame

`dt.year`, `dt.month`, `dt.day` are the in-built attributes to get year, month and day from Pandas datetime object. 

```python
import pandas as pd
```

```python
df = pd.DataFrame({'name':['Ben','Simran','Sanjay'],
                  'DoB':['08-05-1990', '04-28-1998', '12-16-2001']})
```

```python
df
```

```python
df.dtypes
```

```python
df['DoB']=pd.to_datetime(df['DoB'])
```

```python
df.dtypes
```

```python
df['year']= df.DoB.dt.year
df['month']= df.DoB.dt.month
df['day']= df.DoB.dt.day
```

```python
df
```

### Assemble a datetime from multiple columns

```python
df
```

```python
df['is_leap_year?'] = df.DoB.dt.is_leap_year
```
